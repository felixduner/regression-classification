```python
#Task 1
#Linear regression model relating living area to selling price
import numpy as np
import matplotlib.pyplot as plt  
import pandas as pd
#To be able to use linear regression and R2 score
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score

#reading in the data from Hemnet, which we saved in a CSV (col 1 = living area, col 2 = selling price)
df = pd.read_csv('Landvetter_selling_price_living_area.csv')

#creating arrays of the columns. x = living area, y = selling price
x = df.iloc[:, 0].values.reshape(-1, 1)
y = df.iloc[:, 1].values.reshape(-1, 1)

#generating regression model
model = LinearRegression().fit(x,y)

#creating the regression line
xfit = np.linspace(50,220,1000)
yfit = model.predict(xfit[:, np.newaxis])

#plotting
plt.scatter(x,y, color='g')
plt.plot(xfit,yfit, color='r')
plt.xlabel('Living Area (m2)')
plt.ylabel('Selling Price (SEK)')
plt.title('Linear Regression for Selling Price VS Living Area in Landvetter')
plt.show()
```

```python
#1a
#Slope of the line:
print(f"The slope is: {model.coef_[0]}")

#Intercept of the line:
print(f"The intercept is: {model.intercept_[0]}")
```

```python
#1b
#using model.predict to predict prices for 100, 150, 200 m2 houses in Landvetter
print(f"A house of 100 m2 will probably sell for {model.predict([[100]])[0]} SEK")
print(f"A house of 150 m2 will probably sell for {model.predict([[150]])[0]} SEK")
print(f"A house of 200 m2 will probably sell for {model.predict([[200]])[0]} SEK")
```

```python
#1c
#importing seaborn to be able to use residplot()
import seaborn as sb

#plotting the residuals
sb.residplot(x=x, y=y, lowess=True, color="b")
```

```python
#Printing correlation score R2 for model (abs(1)=high linear correlation)
ypred = model.predict(x)
print(f'The correlation score for our model is {r2_score(y,ypred)}')
```
